## 06. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
## 06a. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## 06b. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# creating a new column 'average_price_order' where the average price per order_id is calculated
ords_prods_merge['average_price_order'] = ords_prods_merge.groupby(['order_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge
```

```python
# checking if the calculation makes sense by subsetting for user_id ==1 
ords_prods_merge.loc[ords_prods_merge['user_id']==1]
```

```python
# flagging the new column 'average_price_order'
# mean < 10 = "Low spender"
ords_prods_merge.loc[ords_prods_merge['average_price_order']<10, 'Spender'] = 'Low spender'
```

```python
# mean >= 10 = "High spender"
ords_prods_merge.loc[ords_prods_merge['average_price_order']>=10, 'Spender'] = 'High spender'
```

```python
# checking if it worked
ords_prods_merge['Spender'].value_counts(dropna = False)
```

```python
ords_prods_merge
```

## 07. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
## 07a. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## 07b. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## 07b. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# creating a new column, where I calculate the median of 
# 'days_since_prior_order' per customer
ords_prods_merge['median_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge
```

```python
# changing display option for printing more data sets
pd.options.display.max_rows = 200
```

```python
# checking if it worked
ords_prods_merge['median_order'].value_counts(dropna = False)
```

```python
# subsetting NaN to have understand why there is data missing.
df_nan = ords_prods_merge[ords_prods_merge['median_order'].isnull()]
```

```python
df_nan
```

#### It looks like we have just 5 customers who have ordered 1 time only.

```python
# Why is the max. median 30? Also the maximum days between orders is 30. Why?
ords_prods_merge['days_since_prior_order'].max()
```

```python
# flagging as follows:
# 'median_order' > 20 = 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_order']>20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
# 'median_order' <= 10 = 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_order']<=10, 'customer_frequency'] = 'Frequent customer'
```

```python
# 10 >'median_order' <= 20 = 'Regular customer'
ords_prods_merge.loc[(ords_prods_merge['median_order']>10) & (ords_prods_merge['median_order']<=20), 'customer_frequency'] = 'Regular customer'
```

```python
# checking if it worked
ords_prods_merge['customer_frequency'].value_counts (dropna = False)
```
