```python
# define a function: <=5$ = 'low-range-product', 5-15$ = 'mid-range product', >15$ = 'high-range-product'
def price_label(row):

    if row['prices'] <= 5: 
        return 'Low-range product'
    elif row['prices'] >5 and row['prices'] <=15: 
        return 'Mid-range product'
    elif row['prices'] >15: 
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# we did define the function, we did not apply it to any df
# the difference in starting position of the code is important
```

```python
# use 'price_label' on df
df['price_range'] = df.apply(price_label, axis=1)
```

### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Frequency for 'order_hour_of_day'
df_ords_prods_merged['order_hour_of_day'].value_counts (dropna = False)
```

```python
# my parameter for the new column 'busiest_period_of_day' will be:
# I will use the .loc() function if-statement
# 9-17 (count>=200.000) = "Most orders"
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'] >= 9) & (df_ords_prods_merged['order_hour_of_day'] <= 17), 'busiest_period_of_day'] = 'Most orders'
```

```python
# 0-6 (count<40.000) = "Least orders"
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'] >= 0) & (df_ords_prods_merged['order_hour_of_day'] <= 6), 'busiest_period_of_day'] = 'Least orders'
```

```python
# 23 (count<40.000) = 'Least orders'
df_ords_prods_merged.loc[df_ords_prods_merged['order_hour_of_day'] == 23, 'busiest_period_of_day'] = 'Least orders'
```

```python
# 7-8 (40.000>=count<200.000) = "Average orders"
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'] >= 7) & (df_ords_prods_merged['order_hour_of_day'] <= 8), 'busiest_period_of_day'] = 'Average orders'
```

```python
# 18-22 (40.000>=count<200.000) = "Average orders"
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'] >= 18) & (df_ords_prods_merged['order_hour_of_day'] <= 22), 'busiest_period_of_day'] = 'Average orders'
```
