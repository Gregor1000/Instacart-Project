## 2. Merge dataframes

```python
# merging df.ords with df_ords_prior
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
df_merged_large
```

```python
# checking if the inner merge matched 'both'
# obviously it must, because we chose an INNER Merge
df_merged_large['_merge'].value_counts()
```

```python
# making an outer merge
df_merged_large_outer = df_ords.merge(df_ords_prior, on = ['order_id'], how = 'outer', indicator = True)
```

```python
df_merged_large_outer['_merge'].value_counts()
```
